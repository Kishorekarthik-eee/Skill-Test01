```python
import pandas as pd
data = {
    'Date': ['01-01-2025', '02-01-2025', '03-01-2025', '04-01-2025', '05-01-2025',
             '06-01-2025', '07-01-2025', '08-01-2025', '09-01-2025', '10-01-2025'],
    'Temperature': [15.2, 16.8, 14.5, 13.0, 17.6, 18.1, 16.2, 12.4, 13.8, 15.5],
    'Humidity': [80, 75, 85, 90, 70, 65, 78, 88, 83, 79],
    'Wind Speed': [10.5, 8, 12, 15.2, 7.4, 6, 9.1, 14.3, 11.5, 10],
    'Rainfall': [0, 0, 1.2, 5.4, 0, 0, 0.5, 2.1, 0, 0]
}

df = pd.DataFrame(data)
print(df.isnull().sum())
df.fillna(method='ffill', inplace=True)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Month'] = df['Date'].dt.month
monthly_avg = df.groupby('Month')['Temperature'].mean()
print(monthly_avg)
import matplotlib.pyplot as plt

monthly_avg.plot(kind='bar', color='skyblue')
plt.title('Monthly Average Temperature')
plt.xlabel('Month')
plt.ylabel('Average Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
data = {
    'Student ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice Brown', 'Ben Carter', 'Clara Davis', 'David Evans', 'Eva Foster',
             'Frank Green', 'Grace Hall', 'Henry Ives', 'Isla Jones', 'Jack King'],
    'Maths': [85, 78, 92, 65, 74, 60, 88, 70, 95, 82],
    'Physics': [88, 75, 94, 68, 78, 55, 85, 72, 96, 80],
    'Chemistry': [87, 80, 90, 66, 76, 58, 86, 74, 94, 78]
}

df = pd.DataFrame(data)
print(df[['Maths', 'Physics', 'Chemistry']].describe())
df['Average Score'] = df[['Maths', 'Physics', 'Chemistry']].mean(axis=1)
print(df[['Student ID', 'Name', 'Average Score']])
low_scores = (df[['Maths', 'Physics', 'Chemistry']] < 60).sum(axis=1)
students_below_60 = df[low_scores > 2]
print(students_below_60[['Student ID', 'Name']])
```

```python
# Sample flight data
data = {
    'Flight': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B', 'C', 'A'],
    'Origin': ['NY', 'LA', 'NY', 'NY', 'LA', 'LA', 'NY', 'NY', 'LA', 'NY'],
    'Destination': ['LA', 'NY', 'LA', 'LA', 'NY', 'NY', 'LA', 'LA', 'NY', 'LA'],
    'Departure Delay': [10, -5, 15, 0, 5, -2, 20, 7, -3, 12]
}

df = pd.DataFrame(data)
print(df['Departure Delay'].describe())
avg_delay = df.groupby('Flight')['Departure Delay'].mean()
print(avg_delay)
print("Most common Origin:", df['Origin'].mode()[0])
print("Most common Destination:", df['Destination'].mode()[0])
import matplotlib.pyplot as plt
plt.boxplot(df['Departure Delay'])
plt.title('Distribution of Departure Delays')
plt.ylabel('Delay (minutes)')
plt.show()
```

```python
import numpy as np
point1 = np.array([3, 4])
point2 = np.array([6, 8])
distance = np.linalg.norm(point1 - point2)
print("Euclidean Distance:", distance)
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
dot_product = np.dot(matrix1, matrix2)
print("Dot Product:\n", dot_product)
```

```python
import numpy as np
import pandas as pd
dates = pd.date_range('2025-01-01', periods=365)
temperatures = 20 + 10 * np.sin(np.linspace(0, 3*np.pi, 365)) + np.random.normal(0, 2, 365)
df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})
print(df.describe())
import matplotlib.pyplot as plt
plt.plot(df['Date'], df['Temperature'])
plt.title('Daily Temperature Over a Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# Simulate stock prices
np.random.seed(0)
dates = pd.date_range('2025-01-01', periods=365)
prices = np.cumsum(np.random.normal(0, 1, 365)) + 100
df = pd.DataFrame({'Date': dates, 'Price': prices})
df['MA20'] = df['Price'].rolling(window=20).mean()
df['MA50'] = df['Price'].rolling(window=50).mean()
plt.plot(df['Date'], df['Price'], label='Price')
plt.plot(df['Date'], df['MA20'], label='20-Day MA')
plt.plot(df['Date'], df['MA50'], label='50-Day MA')
plt.title('Stock Prices Over One Year')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
